## Libraries

```python
# Importing all of the necessary packages to read and display data, spatial data, and interactive maps
import pandas as pd
import geopandas as gpd
import matplotlib
import folium
import mapclassify
```

## Data

```python
# Reading in the shelters file
shelters = gpd.read_file("data/shelters.geojson")

# Reading in the Former Municipality Boundaries Data file
fmbd = gpd.read_file('data/Former Municipality Boundaries Data.geojson')
```

### Plotting a bar chart based off of one of the shelter file's columns. 

```python
# Create a new column with the neighbourh column as an interger type
shelters = shelters.astype({'NEIGHBOURH' : 'int'})

# Creating a bart chart with the neighbourhood names as the x-axis and the neighbourh column as the y-axis
shelters.plot(kind="bar", x = "NAME", y = "NEIGHBOURH")
```

### Plotting a static map of the shelter locations.

```python
# Plotting the shelters file as a map
shelters.plot()
```

### Creating an interactive map of the Former Municipality Boundaries Data.

```python
# Plotting an interactive map with the Former Municipality Boundaries Data file
fmbd.explore()
```
